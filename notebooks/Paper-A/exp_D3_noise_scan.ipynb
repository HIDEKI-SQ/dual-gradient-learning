# Exp D3: Noise Rate Scan - Critical Point Detection

## 目的
Non-Oracle二勾配学習の臨界ノイズ率を特定する

### 実験条件
- Noise rates: 45%, 50%, 55%, 60%, 65%, 70%, 75% (7点)
- Clean ratio: 10%
- λ: 0.0, 0.2, 0.3
- Seeds: 0, 1, 2

### runs数: 63 runs
### 推定時間: ~9時間

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
from torch.nn.utils import parameters_to_vector
import torchvision
import torchvision.transforms as transforms
import numpy as np
import json
import time

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
if torch.cuda.is_available():
    torch.backends.cudnn.benchmark = True
    torch.backends.cudnn.deterministic = False
    print(f'GPU: {torch.cuda.get_device_name(0)}')
print(f'Device: {device}')
```

```python
class IndexedDataset(Dataset):
    def __init__(self, dataset):
        self.dataset = dataset
    def __getitem__(self, idx):
        img, label = self.dataset[idx]
        return img, label, idx
    def __len__(self):
        return len(self.dataset)

class BasicBlock(nn.Module):
    expansion = 1
    def __init__(self, in_planes, planes, stride=1):
        super().__init__()
        self.conv1 = nn.Conv2d(in_planes, planes, 3, stride, 1, bias=False)
        self.bn1 = nn.BatchNorm2d(planes)
        self.conv2 = nn.Conv2d(planes, planes, 3, 1, 1, bias=False)
        self.bn2 = nn.BatchNorm2d(planes)
        self.shortcut = nn.Sequential()
        if stride != 1 or in_planes != planes:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_planes, planes, 1, stride, bias=False),
                nn.BatchNorm2d(planes))
    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        out += self.shortcut(x)
        return F.relu(out)

class ResNet18(nn.Module):
    def __init__(self, num_classes=10):
        super().__init__()
        self.in_planes = 64
        self.conv1 = nn.Conv2d(3, 64, 3, 1, 1, bias=False)
        self.bn1 = nn.BatchNorm2d(64)
        self.layer1 = self._make_layer(64, 2, 1)
        self.layer2 = self._make_layer(128, 2, 2)
        self.layer3 = self._make_layer(256, 2, 2)
        self.layer4 = self._make_layer(512, 2, 2)
        self.linear = nn.Linear(512, num_classes)
    def _make_layer(self, planes, num_blocks, stride):
        strides = [stride] + [1]*(num_blocks-1)
        layers = []
        for s in strides:
            layers.append(BasicBlock(self.in_planes, planes, s))
            self.in_planes = planes
        return nn.Sequential(*layers)
    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.layer1(out)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        out = F.adaptive_avg_pool2d(out, 1)
        out = out.view(out.size(0), -1)
        return self.linear(out)
```

```python
BATCH_SIZE = 256
NUM_WORKERS = 4
EPOCHS = 100
LR = 0.1

# D3: ノイズ率スキャン
NOISE_RATES = [0.45, 0.50, 0.55, 0.60, 0.65, 0.70, 0.75]
CLEAN_RATIO = 0.10  # 固定
LAMBDAS = [0.0, 0.2, 0.3]
SEEDS = [0, 1, 2]

total_runs = len(NOISE_RATES) * len(LAMBDAS) * len(SEEDS)
print(f'Total runs: {total_runs}')
print(f'Estimated time: {total_runs * 9.2 / 60:.1f} hours')

def set_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)

def load_cifar10():
    transform_train = transforms.Compose([
        transforms.RandomCrop(32, padding=4),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))])
    transform_test = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))])
    trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
    testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)
    return trainset, testset

def get_data_loaders(trainset, testset):
    indexed_trainset = IndexedDataset(trainset)
    train_loader = DataLoader(indexed_trainset, batch_size=BATCH_SIZE, shuffle=True,
                              num_workers=NUM_WORKERS, pin_memory=True, persistent_workers=True, drop_last=True)
    test_loader = DataLoader(testset, batch_size=BATCH_SIZE, shuffle=False,
                             num_workers=NUM_WORKERS, pin_memory=True, persistent_workers=True)
    return train_loader, test_loader

def inject_noise(labels, noise_rate, seed):
    np.random.seed(seed + 1000)
    noisy_labels = labels.copy()
    n_noisy = int(noise_rate * len(labels))
    noisy_indices = np.random.choice(len(labels), n_noisy, replace=False)
    for idx in noisy_indices:
        choices = [i for i in range(10) if i != labels[idx]]
        noisy_labels[idx] = np.random.choice(choices)
    return noisy_labels

def select_clean_subset(n_samples, clean_ratio, seed):
    np.random.seed(seed + 2000)
    n_clean = int(clean_ratio * n_samples)
    return set(np.random.choice(n_samples, n_clean, replace=False))
```

```python
def evaluate(model, test_loader):
    model.eval()
    correct = total = 0
    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device, non_blocking=True), targets.to(device, non_blocking=True)
            outputs = model(inputs)
            _, predicted = outputs.max(1)
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()
    return correct / total

def train_ce_baseline(model, train_loader, test_loader, noisy_labels):
    optimizer = optim.SGD(model.parameters(), lr=LR, momentum=0.9, weight_decay=5e-4)
    scheduler = optim.lr_scheduler.MultiStepLR(optimizer, milestones=[50, 75], gamma=0.1)
    criterion = nn.CrossEntropyLoss()
    noisy_labels_tensor = torch.tensor(noisy_labels, device=device)
    best_acc = 0
    
    for epoch in range(EPOCHS):
        model.train()
        for inputs, _, indices in train_loader:
            inputs = inputs.to(device, non_blocking=True)
            indices = indices.to(device, non_blocking=True)
            batch_labels = noisy_labels_tensor[indices]
            optimizer.zero_grad()
            loss = criterion(model(inputs), batch_labels)
            loss.backward()
            optimizer.step()
        scheduler.step()
        if (epoch + 1) % 10 == 0:
            best_acc = max(best_acc, evaluate(model, test_loader))
    final_acc = evaluate(model, test_loader)
    return final_acc, max(best_acc, final_acc)

def train_dual_gradient_non_oracle(model, train_loader, test_loader, clean_labels, noisy_labels, clean_indices, lam):
    optimizer = optim.SGD(model.parameters(), lr=LR, momentum=0.9, weight_decay=5e-4)
    scheduler = optim.lr_scheduler.MultiStepLR(optimizer, milestones=[50, 75], gamma=0.1)
    criterion = nn.CrossEntropyLoss()
    clean_labels_tensor = torch.tensor(clean_labels, device=device)
    noisy_labels_tensor = torch.tensor(noisy_labels, device=device)
    K, cached_value_grad, global_step, best_acc = 16, None, 0, 0
    
    for epoch in range(EPOCHS):
        model.train()
        for inputs, _, indices in train_loader:
            inputs = inputs.to(device, non_blocking=True)
            indices_cpu = indices.numpy()
            indices = indices.to(device, non_blocking=True)
            batch_noisy = noisy_labels_tensor[indices]
            
            optimizer.zero_grad()
            loss_struct = criterion(model(inputs), batch_noisy)
            loss_struct.backward()
            g_struct = parameters_to_vector([p.grad for p in model.parameters()]).clone()
            
            if global_step % K == 0 or cached_value_grad is None:
                clean_mask = [i for i, idx in enumerate(indices_cpu) if idx in clean_indices]
                if len(clean_mask) > 0:
                    optimizer.zero_grad()
                    clean_inputs = inputs[clean_mask]
                    clean_targets = clean_labels_tensor[indices[clean_mask]]
                    loss_value = criterion(model(clean_inputs), clean_targets)
                    loss_value.backward()
                    cached_value_grad = parameters_to_vector([p.grad for p in model.parameters()]).clone()
                elif cached_value_grad is None:
                    cached_value_grad = g_struct.clone()
            
            g_struct_norm = g_struct / (g_struct.norm() + 1e-12)
            g_value_norm = cached_value_grad / (cached_value_grad.norm() + 1e-12)
            g_mix = (1 - lam) * g_struct_norm + lam * g_value_norm
            
            optimizer.zero_grad()
            idx = 0
            for p in model.parameters():
                numel = p.numel()
                p.grad = g_mix[idx:idx+numel].view(p.shape).clone()
                idx += numel
            optimizer.step()
            global_step += 1
        scheduler.step()
        if (epoch + 1) % 10 == 0:
            best_acc = max(best_acc, evaluate(model, test_loader))
    final_acc = evaluate(model, test_loader)
    return final_acc, max(best_acc, final_acc)
```

```python
trainset, testset = load_cifar10()
clean_labels = np.array(trainset.targets)
train_loader, test_loader = get_data_loaders(trainset, testset)
print(f'Data loaded.')

print('Verifying IndexedDataset...')
sample_batch = next(iter(train_loader))
print(f'  ✓ Batch: inputs{list(sample_batch[0].shape)}, indices{list(sample_batch[2].shape)}')

print('Warming up GPU...')
warmup_model = ResNet18().to(device)
for _ in range(20):
    _ = warmup_model(torch.randn(BATCH_SIZE, 3, 32, 32, device=device))
del warmup_model
torch.cuda.empty_cache()
print('Warmup complete.')
```

```python
results = []
checkpoint_file = f'{SAVE_DIR}/checkpoint.json'
completed = set()

if os.path.exists(checkpoint_file):
    with open(checkpoint_file, 'r') as f:
        results = json.load(f)
    for r in results:
        completed.add((r['noise_rate'], r['lambda'], r['seed']))
    print(f'Checkpoint loaded: {len(completed)} runs')

run_counter = 0
exp_start = time.time()

for noise_rate in NOISE_RATES:
    print(f'\n{"="*60}')
    print(f'NOISE RATE: {int(noise_rate*100)}%')
    print(f'{"="*60}')
    
    for lam in LAMBDAS:
        for seed in SEEDS:
            run_counter += 1
            key = (noise_rate, lam, seed)
            
            if key in completed:
                print(f'[{run_counter}/{total_runs}] noise={int(noise_rate*100)}% λ={lam} seed={seed} - SKIPPED')
                continue
            
            print(f'\n[{run_counter}/{total_runs}] noise={int(noise_rate*100)}% λ={lam} seed={seed}')
            t0 = time.time()
            
            set_seed(seed)
            noisy_labels = inject_noise(clean_labels, noise_rate, seed)
            clean_indices = select_clean_subset(len(clean_labels), CLEAN_RATIO, seed)
            model = ResNet18().to(device)
            
            if lam == 0.0:
                final_acc, best_acc = train_ce_baseline(model, train_loader, test_loader, noisy_labels)
            else:
                final_acc, best_acc = train_dual_gradient_non_oracle(model, train_loader, test_loader, clean_labels, noisy_labels, clean_indices, lam)
            
            elapsed = time.time() - t0
            
            results.append({
                'experiment': 'exp_D3_noise_scan',
                'noise_rate': noise_rate,
                'clean_ratio': CLEAN_RATIO,
                'lambda': lam,
                'seed': seed,
                'test_acc': final_acc,
                'test_error': 1 - final_acc,
                'best_test_error': 1 - best_acc,
                'time_seconds': elapsed
            })
            
            with open(checkpoint_file, 'w') as f:
                json.dump(results, f, indent=2)
            
            # 判定マーカー
            ce_error = None
            for r in results:
                if r['noise_rate'] == noise_rate and r['lambda'] == 0.0 and r['seed'] == seed:
                    ce_error = r['test_error']
                    break
            
            if ce_error is not None and lam > 0:
                if (1 - final_acc) < ce_error * 0.7:
                    status = ' ✅ IMPROVED'
                elif (1 - final_acc) > ce_error:
                    status = ' ⚠️ DEGRADED'
                else:
                    status = ' ~ marginal'
            else:
                status = ''
            
            print(f'  Error: {1-final_acc:.4f} | Best: {1-best_acc:.4f} | Time: {elapsed/60:.1f} min{status}')

print(f'\n{"="*60}')
print(f'EXPERIMENT COMPLETE')
print(f'Total time: {(time.time()-exp_start)/3600:.2f} hours')
print(f'{"="*60}')
```

```python
import pandas as pd

with open(f'{SAVE_DIR}/exp_D3_results.json', 'w') as f:
    json.dump(results, f, indent=2)

df = pd.DataFrame(results)

print('\n' + '='*60)
print('SUMMARY BY NOISE RATE')
print('='*60)

summary = df.groupby(['noise_rate', 'lambda']).agg({
    'test_error': ['mean', 'std', 'count'],
    'best_test_error': ['mean', 'std']
}).round(4)
print(summary)
summary.to_csv(f'{SAVE_DIR}/exp_D3_summary.csv')

# 臨界点分析
print('\n' + '='*60)
print('CRITICAL POINT ANALYSIS')
print('='*60)

for noise in NOISE_RATES:
    df_n = df[df['noise_rate'] == noise]
    ce_error = df_n[df_n['lambda'] == 0.0]['test_error'].mean()
    dg_error = df_n[df_n['lambda'] == 0.3]['test_error'].mean()
    improvement = (ce_error - dg_error) / ce_error * 100
    status = '✅ EFFECTIVE' if improvement > 20 else ('⚠️ MARGINAL' if improvement > 0 else '❌ DEGRADED')
    print(f'Noise {int(noise*100)}%: CE={ce_error:.3f}, DG={dg_error:.3f}, Δ={improvement:+.1f}% {status}')
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: λ別のノイズ率曲線
ax1 = axes[0]
for lam in LAMBDAS:
    df_l = df[df['lambda'] == lam]
    g = df_l.groupby('noise_rate')['test_error'].agg(['mean', 'std'])
    label = 'CE baseline' if lam == 0.0 else f'Non-Oracle λ={lam}'
    marker = 's' if lam == 0.0 else 'o'
    ax1.errorbar(g.index * 100, g['mean'], yerr=g['std'], marker=marker, capsize=4, 
                 linewidth=2, markersize=8, label=label)

ax1.set_xlabel('Noise Rate (%)', fontsize=12)
ax1.set_ylabel('Test Error', fontsize=12)
ax1.set_title('Non-Oracle Performance vs Noise Rate\n(Clean ratio = 10%)', fontsize=13)
ax1.legend()
ax1.grid(alpha=0.3)
ax1.set_xticks([45, 50, 55, 60, 65, 70, 75])

# Plot 2: 改善率
ax2 = axes[1]
improvements = []
for noise in NOISE_RATES:
    df_n = df[df['noise_rate'] == noise]
    ce_error = df_n[df_n['lambda'] == 0.0]['test_error'].mean()
    dg_error = df_n[df_n['lambda'] == 0.3]['test_error'].mean()
    improvement = (ce_error - dg_error) / ce_error * 100
    improvements.append(improvement)

colors = ['green' if x > 20 else ('orange' if x > 0 else 'red') for x in improvements]
bars = ax2.bar([int(n*100) for n in NOISE_RATES], improvements, color=colors, alpha=0.7, edgecolor='black')
ax2.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
ax2.axhline(y=20, color='green', linestyle='--', alpha=0.5, label='Effective threshold (20%)')
ax2.set_xlabel('Noise Rate (%)', fontsize=12)
ax2.set_ylabel('Improvement over CE (%)', fontsize=12)
ax2.set_title('Error Reduction by Non-Oracle (λ=0.3)\n(Clean ratio = 10%)', fontsize=13)
ax2.legend()
ax2.grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig(f'{SAVE_DIR}/exp_D3_critical_point.png', dpi=300, bbox_inches='tight')
plt.show()

print(f'\nFigure saved to: {SAVE_DIR}/exp_D3_critical_point.png')
```
