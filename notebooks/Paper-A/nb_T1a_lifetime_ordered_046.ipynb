```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torch.nn.utils import parameters_to_vector
import torchvision
import torchvision.transforms as transforms
from torchvision.models import resnet18
import numpy as np

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
if torch.cuda.is_available():
    print(f'GPU: {torch.cuda.get_device_name()}')
```

```python
# Experiment parameters
BATCH_SIZE = 256
NUM_WORKERS = 4
LR = 0.1
K = 16  # Value gradient cache interval

# Lifetime experiment specific
NOISE_RATE = 0.4
INIT_LAMBDA = 0.35           # For ordered initialization
INIT_EPOCHS = 50             # ‚òÖ FIXED: 30‚Üí50 for proper ordered state
INIT_SUCCESS_THRESHOLD = 0.30  # ‚òÖ FIXED: 0.25‚Üí0.30 (realistic threshold)
TARGET_LAMBDA = 0.46         # Target Œª for lifetime measurement
MAX_OBSERVE_EPOCHS = 150     # Maximum observation time (censoring)
COLLAPSE_THRESHOLD = 0.40    # Error threshold for collapse
COLLAPSE_PERSIST = 5         # Epochs to confirm collapse

N_SEEDS = 50

print(f'Œ∑ = {NOISE_RATE}')
print(f'Init: Œª={INIT_LAMBDA} for {INIT_EPOCHS} epochs (success if error ‚â§ {INIT_SUCCESS_THRESHOLD})')
print(f'Target: Œª={TARGET_LAMBDA}, max {MAX_OBSERVE_EPOCHS} epochs')
print(f'Collapse: error‚â•{COLLAPSE_THRESHOLD} for {COLLAPSE_PERSIST} epochs')
print(f'Seeds: {N_SEEDS}')
```

```python
def get_resnet18():
    model = resnet18(weights=None, num_classes=10)
    model.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1, bias=False)
    model.maxpool = nn.Identity()
    return model

class IndexedDataset(Dataset):
    def __init__(self, dataset):
        self.dataset = dataset
    def __getitem__(self, idx):
        img, label = self.dataset[idx]
        return img, label, idx
    def __len__(self):
        return len(self.dataset)

def set_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

def inject_label_noise(labels, noise_rate, seed):
    np.random.seed(seed)
    noisy = labels.copy()
    n_noisy = int(noise_rate * len(labels))
    idx = np.random.choice(len(labels), n_noisy, replace=False)
    for i in idx:
        noisy[i] = np.random.choice([l for l in range(10) if l != labels[i]])
    return noisy

def load_cifar10():
    tr = transforms.Compose([transforms.RandomCrop(32, padding=4), transforms.RandomHorizontalFlip(),
                             transforms.ToTensor(), transforms.Normalize((0.4914,0.4822,0.4465),(0.2023,0.1994,0.2010))])
    te = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.4914,0.4822,0.4465),(0.2023,0.1994,0.2010))])
    return torchvision.datasets.CIFAR10('./data', True, tr, download=True), torchvision.datasets.CIFAR10('./data', False, te, download=True)

def evaluate(model, loader):
    model.eval()
    correct = total = 0
    with torch.no_grad():
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            correct += (model(x).argmax(1) == y).sum().item()
            total += y.size(0)
    return correct / total
```

```python
def train_one_epoch(model, train_loader, opt, sched, clean_t, noisy_t, lam, cached_gv_ref):
    """Train one epoch with dual-gradient method, return avg cos and cached_gv"""
    crit = nn.CrossEntropyLoss()
    model.train()
    step = cached_gv_ref['step']
    cached_gv = cached_gv_ref['gv']
    cos_list = []
    
    for x, _, idx in train_loader:
        x, idx = x.to(device), idx.to(device)
        bn, bc = noisy_t[idx], clean_t[idx]
        
        opt.zero_grad()
        loss_s = crit(model(x), bn)
        loss_s.backward(retain_graph=True)
        gs = parameters_to_vector([p.grad for p in model.parameters()]).clone()
        
        if step % K == 0 or cached_gv is None:
            opt.zero_grad()
            loss_v = crit(model(x), bc)
            loss_v.backward()
            cached_gv = parameters_to_vector([p.grad for p in model.parameters()]).clone()
        
        gs_n = gs / (gs.norm() + 1e-12)
        gv_n = cached_gv / (cached_gv.norm() + 1e-12)
        cos_list.append((gs_n @ gv_n).item())
        
        g_mix = (1 - lam) * gs_n + lam * gv_n
        opt.zero_grad()
        i = 0
        for p in model.parameters():
            n = p.numel()
            p.grad = g_mix[i:i+n].view(p.shape).clone()
            i += n
        opt.step()
        step += 1
    
    if sched is not None:
        sched.step()
    
    cached_gv_ref['step'] = step
    cached_gv_ref['gv'] = cached_gv
    return np.mean(cos_list)
```

```python
def run_lifetime_experiment(seed, train_loader, test_loader, clean_labels, noisy_labels):
    """Run single lifetime experiment: init ‚Üí observe ‚Üí detect collapse"""
    
    clean_t = torch.tensor(clean_labels, device=device)
    noisy_t = torch.tensor(noisy_labels, device=device)
    
    # Phase 1: Initialize to ordered state
    set_seed(seed)
    model = get_resnet18().to(device)
    opt = optim.SGD(model.parameters(), lr=LR, momentum=0.9, weight_decay=5e-4)
    sched = optim.lr_scheduler.MultiStepLR(opt, [50, 75], 0.1)
    
    cached_gv_ref = {'step': 0, 'gv': None}
    init_trajectory = []
    
    print(f'    Phase 1: Initializing (Œª={INIT_LAMBDA}, {INIT_EPOCHS} epochs)...')
    for ep in range(INIT_EPOCHS):
        train_one_epoch(model, train_loader, opt, sched, clean_t, noisy_t, INIT_LAMBDA, cached_gv_ref)
        if (ep + 1) % 10 == 0:
            err = 1 - evaluate(model, test_loader)
            init_trajectory.append({'epoch': ep + 1, 'error': err})
            print(f'      Epoch {ep+1}: error = {err:.4f}')
    
    init_final_error = 1 - evaluate(model, test_loader)
    
    # Check if initialization succeeded
    if init_final_error > INIT_SUCCESS_THRESHOLD:
        print(f'    ‚ö†Ô∏è Init failed: error={init_final_error:.4f} > {INIT_SUCCESS_THRESHOLD}')
        return {
            'seed': seed,
            'init_error': init_final_error,
            'init_success': False,
            'lifetime': 0,
            'censored': False,
            'final_error': init_final_error,
            'init_trajectory': init_trajectory,
            'observe_trajectory': []
        }
    
    print(f'    ‚úÖ Init success: error={init_final_error:.4f}')
    print(f'    Phase 2: Observing (Œª={TARGET_LAMBDA}, max {MAX_OBSERVE_EPOCHS} epochs)...')
    
    # Phase 2: Switch to target Œª and observe
    observe_trajectory = [{'epoch': 0, 'error': init_final_error}]
    collapse_count = 0
    lifetime = None
    
    for ep in range(1, MAX_OBSERVE_EPOCHS + 1):
        train_one_epoch(model, train_loader, opt, sched, clean_t, noisy_t, TARGET_LAMBDA, cached_gv_ref)
        err = 1 - evaluate(model, test_loader)
        observe_trajectory.append({'epoch': ep, 'error': err})
        
        # Check for collapse
        if err >= COLLAPSE_THRESHOLD:
            collapse_count += 1
            if collapse_count >= COLLAPSE_PERSIST:
                lifetime = ep - COLLAPSE_PERSIST + 1  # First epoch of collapse
                print(f'      üíÄ Collapsed at epoch {lifetime} (error={err:.4f})')
                break
        else:
            collapse_count = 0
        
        # Progress report
        if ep % 30 == 0:
            print(f'      Epoch {ep}: error = {err:.4f}')
    
    censored = (lifetime is None)
    if censored:
        lifetime = MAX_OBSERVE_EPOCHS
        print(f'      ‚è±Ô∏è Censored at {MAX_OBSERVE_EPOCHS} epochs')
    
    final_error = 1 - evaluate(model, test_loader)
    
    return {
        'seed': seed,
        'init_error': init_final_error,
        'init_success': True,
        'lifetime': lifetime,
        'censored': censored,
        'final_error': final_error,
        'init_trajectory': init_trajectory,
        'observe_trajectory': observe_trajectory
    }
```

```python
# Load data
trainset, testset = load_cifar10()
clean_labels = np.array(trainset.targets)
train_loader = DataLoader(IndexedDataset(trainset), BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
test_loader = DataLoader(testset, BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

# Warmup
m = get_resnet18().to(device)
for _ in range(10): _ = m(torch.randn(BATCH_SIZE,3,32,32,device=device))
del m; torch.cuda.empty_cache()
print('Ready')
```

```python
# Main experiment loop with checkpoint/resume
results = []
ckpt = f'{SAVE_DIR}/{NOTEBOOK_ID}_checkpoint.json'
done_seeds = set()

if os.path.exists(ckpt):
    results = json.load(open(ckpt))
    done_seeds = {r['seed'] for r in results}
    print(f'Loaded: {len(done_seeds)} done')

for seed in range(N_SEEDS):
    if seed in done_seeds:
        continue
    
    run_num = len(results) + 1
    print(f'\n{"="*50}')
    print(f'[{run_num}/{N_SEEDS}] Seed {seed}')
    print(f'{"="*50}')
    
    # Prepare noisy labels for this seed
    noisy_labels = inject_label_noise(clean_labels, NOISE_RATE, seed)
    
    t0 = time.time()
    result = run_lifetime_experiment(seed, train_loader, test_loader, clean_labels, noisy_labels)
    dt = time.time() - t0
    
    result['experiment_id'] = f'{NOTEBOOK_ID}-{seed:03d}'
    result['target_lambda'] = TARGET_LAMBDA
    result['eta'] = NOISE_RATE
    result['time'] = dt
    
    results.append(result)
    done_seeds.add(seed)
    
    # Summary
    if result['init_success']:
        status = '‚è±Ô∏è CENSORED' if result['censored'] else f"üíÄ COLLAPSED at epoch {result['lifetime']}"
    else:
        status = '‚ùå INIT FAILED'
    print(f'\n  Summary: {status}')
    print(f'  Init error: {result["init_error"]:.4f} | Final error: {result["final_error"]:.4f}')
    print(f'  Time: {dt/60:.1f} min')
    
    # Save checkpoint
    json.dump(results, open(ckpt, 'w'), indent=2)
    
    remaining = N_SEEDS - run_num
    print(f'  ETA: {remaining*dt/3600:.1f}h')
    
    torch.cuda.empty_cache()

print('\n' + '='*50 + f'\n{NOTEBOOK_ID} DONE\n' + '='*50)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Save results (without trajectory for CSV)
json.dump(results, open(f'{SAVE_DIR}/{NOTEBOOK_ID}_results.json', 'w'), indent=2)
df = pd.DataFrame([{k: v for k, v in r.items() if 'trajectory' not in k} for r in results])
df.to_csv(f'{SAVE_DIR}/{NOTEBOOK_ID}_results.csv', index=False)

print('='*60)
print(f'{NOTEBOOK_ID} SUMMARY - LIFETIME MEASUREMENT')
print(f'Target Œª = {TARGET_LAMBDA}, Œ∑ = {NOISE_RATE}')
print('='*60)

# Init success rate
n_init_success = (df['init_success'] == True).sum()
n_init_fail = (df['init_success'] == False).sum()
print(f'\nüìã Initialization:')
print(f'   Success: {n_init_success}/{len(df)} ({100*n_init_success/len(df):.1f}%)')
print(f'   Failed:  {n_init_fail}/{len(df)} ({100*n_init_fail/len(df):.1f}%)')

# Lifetime stats (only for successful inits)
successful = df[df['init_success'] == True]
if len(successful) > 0:
    n_total = len(successful)
    n_collapsed = (successful['censored'] == False).sum()
    n_censored = (successful['censored'] == True).sum()
    
    print(f'\nüìä Lifetime Results (n={n_total}):')
    print(f'   Collapsed: {n_collapsed} ({100*n_collapsed/n_total:.1f}%)')
    print(f'   Censored:  {n_censored} ({100*n_censored/n_total:.1f}%)')
    
    lifetimes = successful['lifetime'].values
    print(f'\n‚è±Ô∏è Lifetime Statistics:')
    print(f'   Mean: {lifetimes.mean():.1f} epochs')
    print(f'   Std:  {lifetimes.std():.1f} epochs')
    print(f'   Min:  {lifetimes.min():.0f} epochs')
    print(f'   Max:  {lifetimes.max():.0f} epochs')
    print(f'   Median: {np.median(lifetimes):.1f} epochs')
```

```python
# Kaplan-Meier survival curve
successful = df[df['init_success'] == True]

if len(successful) > 0:
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    lifetimes = successful['lifetime'].values
    
    # 1. Survival curve
    ax = axes[0]
    times = np.arange(0, MAX_OBSERVE_EPOCHS + 1)
    survival = np.array([np.mean(lifetimes >= t) for t in times])
    
    ax.step(times, survival, where='post', linewidth=2, color='blue')
    ax.fill_between(times, survival, step='post', alpha=0.2, color='blue')
    ax.axhline(0.5, color='red', linestyle='--', alpha=0.7, label='Median')
    ax.set_xlabel('Time (epochs)', fontsize=12)
    ax.set_ylabel('Survival Probability', fontsize=12)
    ax.set_title(f'Survival Curve (Œª={TARGET_LAMBDA})', fontsize=12)
    ax.set_xlim(0, MAX_OBSERVE_EPOCHS)
    ax.set_ylim(0, 1.05)
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # 2. Lifetime histogram
    ax = axes[1]
    collapsed_lifetimes = successful[successful['censored'] == False]['lifetime'].values
    if len(collapsed_lifetimes) > 0:
        ax.hist(collapsed_lifetimes, bins=20, color='red', alpha=0.7, edgecolor='black', label='Collapsed')
    n_censored = (successful['censored'] == True).sum()
    ax.axvline(MAX_OBSERVE_EPOCHS, color='blue', linestyle='--', linewidth=2, label=f'Censored (n={n_censored})')
    ax.set_xlabel('Lifetime (epochs)', fontsize=12)
    ax.set_ylabel('Count', fontsize=12)
    ax.set_title('Lifetime Distribution', fontsize=12)
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # 3. Sample trajectories
    ax = axes[2]
    n_show = min(10, len(results))
    for i, r in enumerate(results[:n_show]):
        if r['observe_trajectory'] and r['init_success']:
            epochs = [t['epoch'] for t in r['observe_trajectory']]
            errors = [t['error'] for t in r['observe_trajectory']]
            color = 'red' if not r['censored'] else 'blue'
            ax.plot(epochs, errors, alpha=0.5, color=color, linewidth=1)
    
    ax.axhline(COLLAPSE_THRESHOLD, color='orange', linestyle='--', linewidth=2, label=f'Collapse ({COLLAPSE_THRESHOLD})')
    ax.axhline(0.20, color='green', linestyle='--', linewidth=2, label='Ordered (0.20)')
    ax.set_xlabel('Epoch (after switch)', fontsize=12)
    ax.set_ylabel('Test Error', fontsize=12)
    ax.set_title(f'Sample Trajectories (n={n_show})', fontsize=12)
    ax.legend(loc='upper right')
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'{SAVE_DIR}/figures/{NOTEBOOK_ID}_survival_analysis.png', dpi=150)
    plt.show()
    
    print(f'\nüìà Figure saved to {SAVE_DIR}/figures/{NOTEBOOK_ID}_survival_analysis.png')
else:
    print('‚ö†Ô∏è No successful initializations - cannot generate survival analysis')
```
